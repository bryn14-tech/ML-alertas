# 00 - Carga y exploración de datos locales

Fase 0: trabajo 100% local. El histórico de alertas (y otras fuentes) se
exportan manualmente como archivos (CSV, Excel o Parquet) dentro de `data/`
(carpeta ignorada por git).

Este notebook lista los archivos disponibles en `data/` y hace una
exploración inicial del histórico de alertas.

```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path("../data")

archivos = sorted(p.relative_to(DATA_DIR) for p in DATA_DIR.rglob("*") if p.is_file() and p.name != ".gitkeep")
archivos
```

## Cargar el histórico de alertas

Ajustar `ARCHIVO_ALERTAS` al nombre real del archivo exportado (CSV, XLSX o
Parquet).

```python
ARCHIVO_ALERTAS = "alertas.csv"  # TODO: ajustar al nombre real del archivo

ruta = DATA_DIR / ARCHIVO_ALERTAS

if ruta.suffix == ".csv":
    df = pd.read_csv(ruta)
elif ruta.suffix in (".xlsx", ".xls"):
    df = pd.read_excel(ruta)
elif ruta.suffix == ".parquet":
    df = pd.read_parquet(ruta)
else:
    raise ValueError(f"Formato no soportado: {ruta.suffix}")

df.shape
```
